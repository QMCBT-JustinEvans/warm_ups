WARMUP: In a notebook, write me a loop FOR numbers 1-100, 
if the number is divisible by 3 print Code, 
if the number is divisible by 5 print Up, 
if it is divisible by 3 and 5 print CodeUp. 
Otherwise just print the number

```python
# program to display numbers from 1 to 100

# initialize the variable
i = 0
n = 100

# while loop from i = 1 to 100
while i <= n:
    i = i + 1
    
    if i == 3:
        print('Code')
        
    if i == 5:
        print('Up')
    
    if i == 35:
        print('Codeup')
        
    else:
        print(i)
```

```python
for i in range(1,101):
    if i% 3 == 0 and i% 5 == 0:
        print('Codeup')
    elif i% 3 == 0:
        print('Code')
    elif i% 5 == 0:
        print('Up')
    else:
        print(i)
```
