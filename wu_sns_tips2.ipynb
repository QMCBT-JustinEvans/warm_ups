```python
encoded_df = pd.DataFrame()
encoded_df['sex_encoded'] = df.sex.map({'Male': 1, 'Female': 0})
encoded_df['smoker_encoded'] = df.smoker.map({'Yes': 1, 'No': 0})
# encoded_df['day_encoded'] = df.day.map({'Thur': 4, 'Fri': 5, 'Sat': 6, 'Sun': 7})
encoded_df['time_encoded'] = df.time.map({'Lunch': 1, 'Dinner': 0})
encoded_cols = encoded_df.columns

df = pd.concat([df, encoded_df], axis=1)

```

```python
df
```

```python
drop_cols = ['sex', 'smoker', 'time'] 
df = df.drop(columns = drop_cols)
```

```python
df
```

```python
dummy_df = pd.get_dummies(data=df[['day']], drop_first=False)
df = pd.concat([df, dummy_df], axis=1)
```

```python
drop_cols = ['day'] 
df = df.drop(columns = drop_cols)
```
